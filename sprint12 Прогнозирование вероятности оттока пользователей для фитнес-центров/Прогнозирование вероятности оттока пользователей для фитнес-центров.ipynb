Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.

Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.
    
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Ваша задача — провести анализ и подготовить план действий по удержанию клиентов.

<font size=+2><b> 3. ПОСТРОЙТЕ МОДЕЛЬ ПРОГНОЗИРОВАНИЯ ОТТОКА КЛИЕНТОВ. </b></font>

При подготовке датафрейма для обучения удалим по одному признаку из пар, которые сильно коррелируют. После построим две модели классификации клиентов: на основании логистической регресии и случайного леса. По метрикам accuracy, precision и recall лучше оказалась модель на основании логистической регрессии. Именно данную модель стоит использовать для реальных прогнозов на новых выборках.
    
Также хотелось бы отметить, что по метрике accuracy получены наибольшие результаты, однако в наших выборках соотношения между классами отлично от равного (не 1:1, а скорее 1:3 в пользу оставшихся клиентов), поэтому данная метрика менее корректна чем другие.
    
Также рассмотрим рассмотрим признаки с наибольшеми весами (по модулю) в уравнении регресии: ими, как и предполагалось выше, оказались 'lifetime', 'contract_period', 'age'. Именно они оказывают наибольшее "влияние" на факт оттока отдельного клиента в нашей модели логистической регрессии.
    
</div>

```python
x = df.drop(columns=['churn', 'month_to_end_contract', 'avg_class_frequency_current_month'])
y = df['churn']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)

x_train_st = scaler.transform(x_train)
x_test_st = scaler.transform(x_test)
```

```python
lr_model = LogisticRegression(solver='liblinear', random_state=0)

lr_model.fit(x_train_st, y_train)

lr_pred = lr_model.predict(x_test_st)
```

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state=0)

rf_model.fit(x_train_st, y_train)

rf_pred = rf_model.predict(x_test_st)
```

```python
print('Метрика accuracy для логистической регрессии и случайного леса:', accuracy_score(y_test, lr_pred), 
      'и', accuracy_score(y_test, rf_pred))

print('\nМетрика precision для логистической регрессии и случайного леса:', round(precision_score(y_test, lr_pred), 3), 
      'и', round(precision_score(y_test, rf_pred), 3))

print('\nМетрика recall для логистической регрессии и случайного леса:', round(recall_score(y_test, lr_pred), 3), 
      'и', round(recall_score(y_test, rf_pred), 3))
```

```python
coef = pd.DataFrame(np.array(lr_model.coef_).transpose(), columns=['weight'], index = x.columns)
coef['abs_weight'] = abs(coef['weight'])

print(coef.sort_values(by='abs_weight', ascending=False))
```
